    In E-commerce, both the retailers (here brands) and the company have to make profit to sustain in the business.
    The E-Commerce company has the following rule for computing their own revenue: </font>
    
####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font>

## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue.

### Ans 5

```python
ECom_Data.head(2)
```

```python
#function to calculate the revenue
def cal_rev(final_price):
  if final_price > 600:
    return (final_price)*(25/100)
  elif final_price>350 and final_price<=650:
    return (final_price)*(15/100)
  elif final_price>100 and final_price<=350:
    return final_price * (10/100)
  else:
    return final_price * (5/100)

ECom_Data['Revenue'] = ECom_Data['discounted_price'].apply(cal_rev) #It will make a column 'Revenue' which will generate profit made by a business after discounted price
ECom_Data.head()
```

```python
Total_Revenue=ECom_Data['Revenue'].sum() #total sum of revenue
print("Total Revenue generated by the E-Commerce company over all orders placed =",Total_Revenue)
```

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
ECom_Data.columns
```

```python
#Order year and month
ECom_Data['Order_Date']=pd.to_datetime(ECom_Data['Order_Date'])
ECom_Data['Order_Month']= ECom_Data['Order_Date'].dt.strftime("%Y-%m")
ECom_Data['Order_Month']
```

```python
#max revenue of each region
max_revenue_by_region = ECom_Data.groupby('Region')['Revenue'].max().reset_index()
max_revenue_by_region
```

```python
#a) Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately
plt.figure(figsize=(20, 7))
plt.title('Monthly Revenue for Different Regions')
plt.xlabel('Month')
plt.ylabel('Revenue')
ECom_Data = ECom_Data.sort_values(by='Order_Month')
#sns.lineplot(data=ECom_Data,x='Month', y='Revenue', hue='Year',errorbar=('ci', False),palette="rainbow")
#sns.lineplot(x='Year', y='Revenue', hue='Month', data=ECom_Data,errorbar=('ci', False))
#sns.lineplot(x='Month', y='Revenue', hue='Region', data=monthly_Revenue, palette='rainbow')
#sns.lineplot(x='Order_Month',y='Revenue',hue='Region',Data=ECom_Data,errorbar=('ci', False))
sns.lineplot(x='Order_Month', y='Revenue',hue ='Region',data=ECom_Data, errorbar=('ci', False))
plt.show()
```

```python
# Group the data by 'Region' and find the maximum monthly revenue and corresponding month for each region
max_revenue_by_region = ECom_Data.groupby('Region').apply(lambda x: x.loc[x['Revenue'].idxmax()]).reset_index(drop=True)
print(max_revenue_by_region[['Region','Revenue','Order_Month' ]])
```

```python
# Group the data by 'Region' and find the minimum monthly revenue and corresponding month for each region
min_revenue_by_region = ECom_Data.groupby('Region').apply(lambda x: x.loc[x['Revenue'].idxmin()]).reset_index(drop=True)
print(min_revenue_by_region[['Region', 'Revenue','Order_Month']])
```
